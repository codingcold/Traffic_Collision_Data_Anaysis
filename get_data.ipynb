```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from opencage.geocoder import OpenCageGeocode
import bs4
from bs4 import BeautifulSoup
import requests
import csv
import seaborn as sns
import urllib
```

```python
def check_link(url):
    try:
        r = requests.get(url)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        print('no link to server!')
rurl = check_link("https://www.splitgraph.com/lacity/traffic-collision-data-from-2010-to-present-d5tf-ez2w")
soup = BeautifulSoup(rurl, 'html.parser')
#print(soup.prettify())
for link in soup.find_all('a'):
    print(link.get('href'))
url = 'https://data.lacity.org/api/views/d5tf-ez2w/rows.csv?accessType=DOWNLOAD'
#f=requests.get(url)
#with open("test.csv","w") as code:
     #code.write(f.content)
testfile = urllib.request.urlopen(url)
file = pd.read_csv(testfile)
file.to_csv('Traffic_Collision_Data_from_2010_to_Present.csv' , index = False , sep = ',')
print(file)
```

```python
location = []
age = []
time = []
latitudes = []
longitudes = []
#read directly
df = pd.read_csv("Traffic_Collision_Data_from_2010_to_Present.csv")
for i in df["Victim Age"].head(10000):
    if i > 0:
        age.append(i)
        location.append(eval(df["Location"].head(10000)[i]))
        time.append((df["Time Occurred"].head(10000)[i]))
for i in location:
    latitudes.append(i[0])
    longitudes.append(i[1])
dataframe = pd.DataFrame({'Victim Age' : age , 'Time Occurred' : time , 'Lat' : latitudes , 'Lng' : longitudes})
print(dataframe)
dataframe.to_csv('raw_data.csv' , index = False , sep = ',')
raw = pd.read_csv('raw_data.csv')
```

```python
lats = raw['Lat'].to_list()
lons = raw['Lng'].to_list()
address = []
key = "b710f1022afa4f6d8fa2a153b1dd9891"
geocoder = OpenCageGeocode(key)
for lat,lng in zip(lats , lons):
    result = geocoder.reverse_geocode(lat,lng)
    print(result[0]['formatted'])
    address.append(result[0]['formatted'])
```

```python
print(len(address))
```

```python
import re
zipco = []
for i in address:
    phone_pattern = i
    phone_pattern = re.compile(r'(?<=CA )\d+')
    a = phone_pattern.findall(i)
    if a != []:
        zipco.append(a[0])
print(len(zipco))
```

```python
population_count = pd.read_csv("population.csv")
zipcode = []
for i in  population_count["Zip Code"]:
    zipcode.append(i)
print(len(zipcode))
population = []
for j in population_count["Total Population"]:
    population.append(j)
#print(len(population))
data3 = {}
k = 0
while k < len(zipcode):
    data3[str(zipcode[k])] = population[k]
    k += 1
result = {}
for key in zipco:
    result[key] = result.get(key, 0) + 1
print(result)
print(list(result.keys()))
import pandas as pd
import numpy as np
pop = []
for i in list(result.keys()):
    pop.append(data3[i])
dataframe = pd.DataFrame({'zipcode' : list(result.keys()) , 'frequency' : list(result.values()) , 'population' : pop})
dataframe.to_csv('frequency.csv' , index = False , sep = ',')
```
